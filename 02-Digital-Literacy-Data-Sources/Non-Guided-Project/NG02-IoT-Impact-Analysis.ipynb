## The Devices

You're evaluating these 6 smart home devices:

1. **Amazon Echo Dot** - Voice assistant for music, questions, smart home control
2. **Ring Video Doorbell** - See and talk to visitors from your phone
3. **Nest Smart Thermostat** - Automatically adjusts temperature, saves energy
4. **Philips Hue Smart Lights** - Control lights from phone, set schedules
5. **Samsung Smart TV** - Stream content, voice control, apps
6. **August Smart Lock** - Lock/unlock door remotely, give temporary access

For each device, you have data about:
- **Monthly energy cost** (how much it adds to your electric bill)
- **Privacy rating** (1 = minimal data collection, 5 = extensive data collection)
- **Utility rating** (1 = not very useful, 5 = very useful)

## Step 1: The Data (Already Done For You!)

Run this cell to create the data. You don't need to modify anything here.

```python
# Device names
devices = [
    "Echo Dot",
    "Ring Doorbell",
    "Smart Thermostat",
    "Smart Lights",
    "Smart TV",
    "Smart Lock"
]

# Monthly energy cost in dollars
monthly_costs = [
    0.50,   # Echo Dot
    2.00,   # Ring Doorbell
    8.00,   # Smart Thermostat
    3.50,   # Smart Lights
    12.00,  # Smart TV
    1.50    # Smart Lock
]

# Privacy rating: 1 (minimal data) to 5 (extensive data collection)
privacy_ratings = [
    4,  # Echo Dot - always listening
    5,  # Ring Doorbell - video/audio recording
    3,  # Smart Thermostat - temperature and schedule data
    2,  # Smart Lights - just on/off times
    4,  # Smart TV - viewing habits, voice commands
    3   # Smart Lock - entry/exit times
]

# Utility rating: 1 (not very useful) to 5 (very useful)
utility_ratings = [
    5,  # Echo Dot - very versatile
    4,  # Ring Doorbell - security benefit
    5,  # Smart Thermostat - saves money long-term
    3,  # Smart Lights - convenient but not essential
    4,  # Smart TV - entertainment hub
    3   # Smart Lock - convenient but traditional locks work
]

print("Data loaded successfully!")
print(f"\nAnalyzing {len(devices)} smart home devices...")
```

---

## Step 2: Display the Data

Let's print out all the information in a nice format so you can see what we're working with.

**Your task:** Complete the for loop to display each device with its information.

```python
print("="*60)
print("SMART HOME DEVICE ANALYSIS")
print("="*60)

print("\nDevice Details:")
print("-"*60)

# TODO: Complete this loop to print each device's information
# HINT: Use range(len(devices)) to loop through all devices
# HINT: Use index i to access items from all lists

for i in range(len(devices)):
    # TODO: Print device name, cost, privacy rating, and utility rating
    # Format: "Device: Echo Dot | Cost: $0.50 | Privacy: 4 | Utility: 5"
    print(f"Device: {devices[i]}")
    print(f"  Monthly Cost: ${monthly_costs[i]:.2f}")
    print(f"  Privacy Rating: {privacy_ratings[i]}/5")
    print(f"  Utility Rating: {utility_ratings[i]}/5")
    print()

print("-"*60)
```

### Expected Output:

You should see all 6 devices listed with their costs, privacy ratings, and utility ratings.

## Milestone 1: Calculate Total Monthly Energy Cost

If you buy all 6 devices, how much will they add to your monthly electric bill?

**Your task:** Calculate the total monthly cost.

```python
print("\nðŸ’° ENERGY COST ANALYSIS")
print("="*60)

# TODO: Calculate the total monthly cost
# HINT: Use the sum() function on the monthly_costs list
total_monthly_cost = sum(monthly_costs)

# TODO: Calculate the total yearly cost
# HINT: Multiply monthly cost by 12
total_yearly_cost = total_monthly_cost * 12

# Print the results
print(f"Total monthly energy cost: ${total_monthly_cost:.2f}")
print(f"Total yearly energy cost: ${total_yearly_cost:.2f}")

# TODO: Calculate average cost per device
# HINT: Divide total by number of devices
average_cost = total_monthly_cost / len(devices)
print(f"Average cost per device: ${average_cost:.2f}/month")
```

### âœ… Check Your Work:

Your total monthly cost should be around **$27.50**

---

## Milestone 2: Find Most and Least Expensive Devices

Which device costs the most to run? Which costs the least?

**Your task:** Find the highest and lowest cost devices.

```python
print("\nðŸ“Š COST COMPARISON")
print("="*60)

# TODO: Find the highest monthly cost
# HINT: Use the max() function
highest_cost = max(monthly_costs)

# TODO: Find the lowest monthly cost
# HINT: Use the min() function
lowest_cost = min(monthly_costs)

# TODO: Find the INDEX of the most expensive device
# HINT: Use .index() method on the list
most_expensive_index = monthly_costs.index(highest_cost)

# TODO: Find the INDEX of the least expensive device
least_expensive_index = monthly_costs.index(lowest_cost)

# Print the results
print(f"Most expensive device: {devices[most_expensive_index]}")
print(f"  Cost: ${highest_cost:.2f}/month")
print()
print(f"Least expensive device: {devices[least_expensive_index]}")
print(f"  Cost: ${lowest_cost:.2f}/month")
print()
print(f"Cost difference: ${highest_cost - lowest_cost:.2f}/month")
```

### âœ… Check Your Work:

- **Most expensive:** Smart TV ($12.00/month)
- **Least expensive:** Echo Dot ($0.50/month)

## Milestone 3: Privacy Analysis

How much data are these devices collecting? Let's analyze the privacy concerns.

**Your task:** Calculate average privacy rating and identify devices with high privacy concerns.

```python
print("\nðŸ”’ PRIVACY ANALYSIS")
print("="*60)

# TODO: Calculate the average privacy rating
# HINT: Use sum() and len()
total_privacy = sum(privacy_ratings)
average_privacy = total_privacy / len(privacy_ratings)

print(f"Average privacy rating: {average_privacy:.1f}/5")
print()

# TODO: Count how many devices have HIGH privacy concerns (rating >= 4)
# HINT: Use a for loop and a counter variable
high_privacy_concern = 0

for rating in privacy_ratings:
    if rating >= 4:
        high_privacy_concern += 1

print(f"Devices with HIGH privacy concerns (4-5): {high_privacy_concern}")
print()

# Print which devices have high privacy concerns
print("Devices collecting extensive data:")
for i in range(len(devices)):
    if privacy_ratings[i] >= 4:
        print(f"  - {devices[i]} (Rating: {privacy_ratings[i]}/5)")
```

### âœ… Check Your Work:

You should find **3 devices** with high privacy concerns (Echo Dot, Ring Doorbell, Smart TV)

---

## Milestone 4: Utility Analysis

Are these devices actually useful? Let's see which ones provide the most value.

**Your task:** Analyze utility ratings and find the most/least useful devices.

```python
print("\nâ­ UTILITY ANALYSIS")
print("="*60)

# TODO: Calculate average utility rating
average_utility = sum(utility_ratings) / len(utility_ratings)
print(f"Average utility rating: {average_utility:.1f}/5")
print()

# TODO: Find the most useful device
# HINT: Use max() and .index()
highest_utility = max(utility_ratings)
most_useful_index = utility_ratings.index(highest_utility)

print(f"Most useful device: {devices[most_useful_index]}")
print(f"  Utility rating: {highest_utility}/5")
print()

# TODO: Count devices with HIGH utility (rating >= 4)
high_utility_count = 0
for rating in utility_ratings:
    if rating >= 4:
        high_utility_count += 1

print(f"Highly useful devices (4-5 rating): {high_utility_count}")
print()

# List the highly useful devices
print("Top-rated devices:")
for i in range(len(devices)):
    if utility_ratings[i] >= 4:
        print(f"  - {devices[i]} (Rating: {utility_ratings[i]}/5)")
```

### âœ… Check Your Work:

You should find **4 devices** with high utility ratings.

**Note:** If there are multiple devices with rating 5 (Echo Dot and Smart Thermostat), Python will find the first one.

---

## Milestone 5: Final Recommendations

Based on your analysis, which devices offer the best value?

**Your task:** Identify "best value" devices - high utility, low privacy concerns, reasonable cost.

```python
print("\nðŸ’¡ RECOMMENDATIONS")
print("="*60)

print("\nBest Value Devices (High utility, Low privacy concerns):")
print()

# TODO: Find devices that are:
# - Utility rating >= 4 (useful)
# - Privacy rating <= 3 (low privacy concerns)
# HINT: Use a for loop with multiple conditions

for i in range(len(devices)):
    if utility_ratings[i] >= 4 and privacy_ratings[i] <= 3:
        print(f"âœ… {devices[i]}")
        print(f"   Utility: {utility_ratings[i]}/5 | Privacy: {privacy_ratings[i]}/5 | Cost: ${monthly_costs[i]:.2f}/month")
        print()

print("\nDevices to Reconsider (High privacy concerns):")
print()

# TODO: List devices with privacy rating >= 4
for i in range(len(devices)):
    if privacy_ratings[i] >= 4:
        print(f"âš ï¸  {devices[i]}")
        print(f"   Privacy: {privacy_ratings[i]}/5 | Utility: {utility_ratings[i]}/5")
        print()
```

### âœ… Check Your Work:

**Best value device:** Smart Thermostat (high utility, moderate privacy, saves energy long-term)

**Devices to reconsider:** Echo Dot, Ring Doorbell, Smart TV (all have privacy rating 4 or higher)

## Complete Summary Report

Let's create a final summary of all our findings.

```python
print("\n" + "="*60)
print("COMPLETE SMART HOME ANALYSIS SUMMARY")
print("="*60)

print("\nðŸ“Š ENERGY COSTS")
print(f"   Total monthly cost: ${sum(monthly_costs):.2f}")
print(f"   Total yearly cost: ${sum(monthly_costs) * 12:.2f}")
print(f"   Most expensive: {devices[monthly_costs.index(max(monthly_costs))]} (${max(monthly_costs):.2f}/month)")
print(f"   Least expensive: {devices[monthly_costs.index(min(monthly_costs))]} (${min(monthly_costs):.2f}/month)")

print("\nðŸ”’ PRIVACY")
print(f"   Average privacy rating: {sum(privacy_ratings)/len(privacy_ratings):.1f}/5")
high_privacy = sum(1 for rating in privacy_ratings if rating >= 4)
print(f"   Devices with high privacy concerns: {high_privacy}")

print("\nâ­ UTILITY")
print(f"   Average utility rating: {sum(utility_ratings)/len(utility_ratings):.1f}/5")
high_utility = sum(1 for rating in utility_ratings if rating >= 4)
print(f"   Highly useful devices: {high_utility}")

print("\nðŸ’¡ BOTTOM LINE")
print("   Best overall value: Smart Thermostat")
print("     â€¢ High utility (saves energy long-term)")
print("     â€¢ Moderate privacy concerns")
print("     â€¢ Pays for itself through energy savings")

print("\n" + "="*60)
```
